# Nominal /OHE Encoding
One hot encoding,also known as nominal encoding,is a technique used to present categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit correseponds to a unique category.
For example, if we have a categorical variable "color" with three possible values(red,green,blue), we can represent it as one hot encoding as follows:

1.Red:[1,0,0]<br>
2.Green:[0,1,0]<br>
3.Blue:[0,0,1]<br>

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
## Create an instance of OneHotEncoder
encoder=OneHotEncoder()
```

```python
## Perform fit and transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
## for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

# Label Encoding
Label Encoding and odinal encoding are two techniques used to encode categorical data as numerical data.

Label encoding involves assigning a unique numerical label to each category in the variable. The labels are usually assigned in alphabetical order or based on the frequency of the categories. For example, if we have categorical variable "color" with three possible values(red,green,blue), we can represent it using label encoding as follows:

1.Red:1<br>
2.Green:2<br>
3.Blue:3<br>

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['green']])
```

### Odinal Encoding

It is used to encode categorical data that have an intrinsic order or ranking. For example, if we have a categorical variable "education level" with four possible values (high school, college,graduate), we can represent it using ordinal encoding as follows:<br>
1.High school:1 <br>
2.College:2<br>
3:Graduate:3<br>
4:Post-Graduate:4<br>

```python
## Odinal Encoding
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
# Creating a sample DataFrame with an odinal variable
df = pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
    })
```

```python
df
```

```python
# Create an instance of Ordinal Encoder and then fit_transform
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```
